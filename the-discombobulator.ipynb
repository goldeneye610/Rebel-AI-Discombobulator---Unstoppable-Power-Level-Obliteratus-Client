## Step 4: Launch OBLITERATUS Gradio Web UI

The UI will appear **in the output cell below** after you run it.

**Important:** Gradio in Colab uses a public URL like `https://xxxx.gradio.live`. Your session is secured by Google Colab's auth; however, avoid exposing sensitive models publicly.

**Tips for running an ablation:**
1. In the UI, go to the **Models** tab → search for a model (e.g., `mistralai/Mistral-7B-Instruct-v0.3`)
2. Click **Load** → it will download the model into Colab's VM (~5–10 min for 7B)
3. Go to **Obliterate** → choose method `advanced` (default)
4. Set output directory: `./abliterated-models`
5. Click **Run** — it will run and show live progress
6. When done, go to **Models** tab, browse `./abliterated-models`, and download your result


````python
# Step 4: Launch the Gradio UI
print("[+]  OBLITERATUS Web UI on Colab…")
print("The UI link will appear in the output below (usually takes 10–30 seconds)")
print("")

import subprocess
import sys

# Launch obliteratus UI
proc = subprocess.Popen(
    [sys.executable, '-m', 'obliteratus.cli', 'ui', '--port', '7860'],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    cwd='/content/OBLITERATUS'
)

# Print startup logs for a few seconds
import time
for _ in range(20):
    line = proc.stdout.readline()
    if line:
        print(line.decode('utf-8', errors='replace').rstrip())
    time.sleep(1)
    if b'Running on' in line or b'gradio' in line.lower():
        break

print("[✓] Gradio server started! Look for the public URL above (~10–30 sec)")

# Keep the cell alive so the server continues running
try:
    while True:
        time.sleep(1)
except KeyboardInterrupt:
    proc.terminate()
    print("[!] UI terminated by operator.")
````
